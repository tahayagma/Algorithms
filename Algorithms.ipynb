## Binary Search Algorithms

Sıralanmış bir dizide hedef değerin bulunması işlemidir. Dizi sıralı değilse sıralı hale getirilir. Ardından tam ortadaki değer alınır ve hedef değer ile karşılaştırılır. Eğer büyükse dizinin büyük kısmında eğer değilse küçük kısmında arama aynı mantıkla devam eder.

```python
def dizi_sırala(dizi):
    pivot = dizi[0]
    counter = 0
    m = len(dizi)
    c = m
    if m%2 == 0:
        c = m+1
    while counter<c:
        for i in range(1,m):
            if dizi[i]<pivot:
                temp = dizi[i-1]
                dizi[i-1] =dizi[i]
                dizi[i] = temp
                #pivot = dizi[i]
            pivot = dizi[i]
        counter+=1
    return dizi

def bs(dizi,hedef):
    s_dizi = dizi_sırala(dizi)
    bas = 0
    son = len(s_dizi)-1
    sonuc = -1
    while bas<=son:
        orta = (bas+son)//2
        if s_dizi[orta]>hedef:
            son = orta-1
        elif s_dizi[orta]<hedef:
            bas = orta+1
        else:
            print("Bulundu")
            sonuc = orta
            break
    return sonuc
```

## BFS( Breadth First Search) Geniş Önceliklik Arama

```python
class kuyruk():
    
    def __init__(self):
        self.kuyruk = [] # boş bir liste tanımlandı
        self.bas = 0 # kuyrukta ilk değeri almak için 
        print("kuyruk olustu...\n")
        
    def ekle(self,deger):
        if type(deger) == list or type(deger) == tuple:
            for i in deger:
                self.kuyruk.append(i)
        else:
            self.kuyruk.append(deger)
        
    def display(self):
        if self.bos_mu():
            print(self.kuyruk)
        
    def sil(self):
        if self.bos_mu(): # eğer kuyruk boş değilse
            print("silinen: ",self.kuyruk[self.bas]) # silinecek değeri göster
            self.kuyruk= self.kuyruk[self.bas+1:] # silinecek değeri at ve kuyruğu güncelle
        
    def bos_mu(self):
        b = True
        if len(self.kuyruk) == 0:
            b = False
        return b
```

```python
class BFS():
    
    def __init__(self, nodes):
        self.g = graph(nodes)
        self.k = kuyruk() # kuyruk sınıfından k adında bir nesne türetildi
        self.ilk = list(self.g.keys())[0] # grafın ilk düğümü çekildi.
        self.k.ekle(self.ilk) # grafın ilk düğümü kuyruğa eklendi.
        print(f"Kuyruk yapısı oluşturuldu {self.ilk} düğümü kuyruğa eklendi.")
        self.visited = [] # ziyaret edilen düğümleri tutmak için boş bir liste tanımlandı
        self.traverse(self.g,self.k) # graf ve kuyruk yapıları traverse fonksiyonuna gönderildi
    def traverse(self,g,k):
        while self.k.bos_mu(): # önce kuyruğun boş olup olmadığı sorgulandı.
            for i in self.k.kuyruk: # kuyruk üzerinde gezin
                self.yavrular = g[i] # kuyruktan alınan düğümün komşularını aL
                if i in self.visited:
                    self.k.sil()
                else:
                    for x in self.visited:
                        if x in self.yavrular:
                            self.yavrular.remove(x)

                    print(f"{i} düğümü gezildi ve kuyruktan silindi {self.yavrular} kuyruğa eklendi")
                    self.k.ekle(self.yavrular)
                    self.k.sil()
                    self.visited.append(i)
        print("sırasıyla ziyaret edilen düğümler: ",self.visited)
        
```

## Quick Sort

```python
def q(dizi,bas,son):
    pivot = dizi[son] # pivot degerini dizinin sonundan seçtik
    i = (bas-1) # i degeri de -1 değeri alıyor
    for j in range(bas,son): # baştan sona dizi üzerinde gez
        if dizi[j]<pivot: # eger dizinin j. değeri pivottan küçükse
            i+=1 # i değeri bir artıp 
            dizi[i],dizi[j] = dizi[j],dizi[i] # dizinin i. ve j. değerleri swap edilir.
    dizi[i+1],dizi[son] = pivot,dizi[i+1] # eğer j değeri pivota eşit olursa i. değerin bir sonraki değeri ile pivot swap et
    return (i+1)

def s(dizi,bas,son):
    if bas<son:
        i = q(dizi,bas,son) 
        # burda q fonksiyonundan dönen i değeri ile sağ ve sol dizilerinin sınırları belirlenip tekrar s fonksiyonuna
        # gönderiliyor
        
        s(dizi,bas,i-1) #  pivottan küçük değerler
        s(dizi,i+1,son) # pivottan büyük değerler
    return dizi
```

```python
def Graphs(dizi):
    graph = {}
    subdict ={}
    for i in dizi:
        s = input(f"{i} düğümünün komşularını girin(Arada boşluk bırak): ")
        a = s.split(" ")
        for x in a:
            num = input(f"{i}-{x} maliyeti: ")
            subdict[x] = num
        if subdict == {}:
            pass
        else:
            graph[i] = subdict
            subdict = {}
    return graph

def sort_edge(graph):
    d = {}
    for i in graph.keys():
        k = graph[i]
        for x in k:
            a = f"{i}-{x}"
            b = f"{x}-{i}"
            if b  not in d.keys():
                d[a] = int(graph[i][x])
    #print(d)
    return dict(sorted(d.items(),key= lambda x: x[1])[:-1])

def sort_path(sorted_edge,graph):
    visited = {}
    for x in graph.keys():
        visited[x] = -1
    n = []
    counter = 1
    toplamMaliyet = 0
    for i in sorted_edge.keys():
        a = i.split("-")
        # kimden ve kime değerleri 1'e eşit değilse 1 başlayarak değerler atıyoruz.'''
        if visited[a[0]] != 1 and visited[a[1]] !=1:
            visited[a[0]] = counter
            visited[a[1]] = counter
            counter+=1
            n.append(i)
            toplamMaliyet += sorted_edge[i]
        # eğer kimden veya kime değerlerinden herhangi birisi -1 ise yani bunları 1 olarak işaretliyoruz yani gezildi diye
        elif visited[a[0]] == -1 or visited[a[1]] ==-1:
            visited[a[0]] = 1
            visited[a[1]] = 1
            n.append(i)
            toplamMaliyet += sorted_edge[i]
        # eğer kimden veya kime değerlerinden biri 1'den büyükse 1'e eşitleyip daha önceki 1'den büyükleri de 1'e eşitleniyor
        elif visited[a[0]]>1 or visited[a[1]]>1:
            for c in visited.keys():
                if visited[c]>1:
                    visited[c] = 1
            visited[a[0]] = 1
            visited[a[1]] = 1
            n.append(i)
            toplamMaliyet += sorted_edge[i]
        else:pass
    print("en kısa yol: ",n)
    print("toplam maliyet: ",toplamMaliyet)
    #print("visited: ",visited)
```

## Insertion Sort Algorithm

```python
def Insert_Sort(dizi):
    counter = 0
    while counter<len(dizi):
        pivot = dizi[0]
        for i in range(1,len(dizi)):
            if pivot>dizi[i]:
                temp = dizi[i]
                dizi[i-1] = temp
                dizi[i] = pivot
            pivot = dizi[i]
        counter+=1
    return dizi
```

##  Selection sort Algorithm

```python
def select_sort(dizi):
    counter = 0
    while counter<len(dizi):
        pivot= dizi[counter]
        idxmin = 0
        for i in range(counter,len(dizi)):
            if pivot>dizi[i]:
                pivot = dizi[i]
                idxmin = i
        if idxmin != 0:
            dizi[idxmin],dizi[counter] = dizi[counter],pivot
        counter+=1
    return dizi
```

## Counting Sort Algorithm

```python
def count_sort(dizi):
    indis = [0]*(max(dizi)+1)
    for i in dizi:
        indis[i] +=1
    counter = 0
    for i in range(len(indis)):
        while indis[i]>0:
            dizi[counter] = i
            indis[i] -=1
            counter+=1
    return indis,dizi
```

## Quick Select Algoritması

Parçalar ve yönet mantığı ile çalışır.

```python
def partition(dizi,bas,son):
    pivot = dizi[bas]
    i,j = bas+1,bas+1
    while j<len(dizi):
        if dizi[j]<pivot:
            temp = dizi[i]
            dizi[i] = dizi[j]
            dizi[j] = temp
            i+=1
        j+=1
    dizi[i-1],dizi[bas] = pivot,dizi[i-1]
    return i-1

def quickselect(dizi,k,bas,son):
    s = partition(dizi,bas,son)
    if s>k-1:
        return quickselect(dizi,k,bas,s+1)
    elif s<k-1:
        return quickselect(dizi,k,s+1,son)
    else:
        return dizi[k-1]
```

## Boyer–Moore Majority Vote Algorithm(Çoğunluk Oylaması Algoritması)

```python
def majority(nums):
    d = {}
    for i in nums:
        d[i] = d.get(i,0)+1
    for n,f in d.items():
        if f>(len(nums)/2):
            print("The majority element is : ",n)
    return d
majority([1, 8, 7, 4, 1, 2, 2, 2, 2, 2,2])
```

```python
def findMajorityElement(nums):
    m = -1
    i = 0
    for j in range(len(nums)):
        if i == 0:
            m = nums[j]
            i = 1
        elif m == nums[j]:
            i = i + 1
        else:
            i = i - 1
    return m

findMajorityElement([1, 8, 7, 4, 1, 2, 2, 2, 2,4,4,4,4,4,4])
```
